```python
import pandas as pd
import joblib
from sklearn.metrics import r2_score
```

### Importing Model and scaler 

```python
rf = joblib.load('../models/rf_model.pkl')
scaler = joblib.load('../models/scaler.pkl')
```

### Importing Data from Huggingface

```python
df_generated_v1 = pd.read_parquet("hf://datasets/kinisi/gym_kr2-PickYCB-v1_generated/data/chunk-000/episode_000000.parquet")
df_generated_v2 = pd.read_parquet("hf://datasets/kinisi/gym_kr2-PickYCB-v1_generated_v2/data/chunk-000/episode_000000.parquet")
```

```python
v1_state = pd.DataFrame(df_generated_v1['observation.state'].tolist())
v1_state.columns = [f'state_{i}' for i in range(v1_state.shape[1])]
v1_action = df_generated_v1['action'].tolist()


v2_state = pd.DataFrame(df_generated_v2['observation.state'].tolist())
v2_state.columns = [f'state_{i}' for i in range(v2_state.shape[1])]
v2_action = df_generated_v2['action'].tolist()
```

### Standardising observation state

```python
v1_scaled = scaler.transform(v1_state)
v2_scaled = scaler.transform(v2_state)
```

### Predicting action on generated_v1 dataset

```python
v1_pred = rf.predict(v1_scaled)
v1_score = r2_score(v1_action, v1_pred)
print(v1_score)
```

### Predicting action on generated_v2 dataset 

```python
v2_pred = rf.predict(v2_scaled)
v2_score = r2_score(v2_action, v2_pred)
print(v2_score)
```
