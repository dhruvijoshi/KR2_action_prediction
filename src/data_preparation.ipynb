```python
import pandas as pd
```

# Loading Datasets

```python
df_generated = pd.read_parquet("hf://datasets/kinisi/gym_kr2-PickYCB-v1-fixed-depth_generated_v2.1/data/chunk-000/episode_000000.parquet")
df_fake_generated = pd.read_parquet("hf://datasets/kinisi/gym_kr2-PickYCB-v1-fixed-depth_fake_generated_v2.1/data/chunk-000/episode_000000.parquet")
```

```python
df_generated.info()
```

```python
df_fake_generated.info()
```

# New Dataframes for 'observation.state' and 'action'

```python
state_combined = pd.concat([df_generated['observation.state'], df_fake_generated['observation.state']], ignore_index=True)
action_combined = pd.concat([df_generated['action'], df_fake_generated['action']], ignore_index=True)
```

```python
state_df = pd.DataFrame(state_combined.tolist())
state_df.columns = [f'state_{i}' for i in range(state_df.shape[1])]

action_df = pd.DataFrame(action_combined.tolist())
action_df.columns = [f'state_{i}' for i in range(action_df.shape[1])]
```

```python
state_df.to_csv('../data/observation_data', index=False)
action_df.to_csv('../data/action_data', index=False)
```
