```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import joblib
```

```python
X = pd.read_csv('../data/observation_data')
y = pd.read_csv('../data/action_data')
```

```python
X.head()
```

```python
X.describe().round(3)
```

# Spliting data into Train and Test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.describe().round(3)
```

```python
X_test.describe().round(3)
```

# Standardising Data

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=[f'state_{i}' for i in range(X_train_scaled.shape[1])])

X_test_scaled = scaler.transform(X_test)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=[f'state_{i}' for i in range(X_test_scaled.shape[1])])
```

# Model Training

### SKlearn random forest 

```python
rf = RandomForestRegressor(random_state=12)
rf.fit(X_train_scaled, y_train)
y_pred = rf.predict(X_test_scaled)
score = r2_score(y_test, y_pred)
```

```python
score
```

### Saving model and scaler 

```python
joblib.dump(scaler, '../models/scaler.pk1')
joblib.dump(rf, '../models/rf_model.pk1')
```
